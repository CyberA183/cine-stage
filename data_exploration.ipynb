```python
movies_latest = movies[movies['year'] >= 1980].reset_index()
movies_latest
```

```python
#Simplifying genres column
movies_latest['genres'].astype(str)
movies_latest['genres'] = movies_latest['genres'].str.replace('[', '', regex=False).str.replace(']', '', regex=False).str.replace("'", '', regex=False)
movies_latest['genres']
```

```python
#Simplifying production companies column
movies_latest['production_companies'].astype(str)
movies_latest['production_companies'] = movies_latest['production_companies'].str.replace('[', '', regex=False).str.replace(']', '', regex=False).str.replace("'", '', regex=False)
movies_latest['production_companies']
```

```python
#Simplifying language column
movies_latest['languages'].astype(str)
movies_latest['languages'] = movies_latest['languages'].str.replace('[', '', regex=False).str.replace(']', '', regex=False).str.replace("'", '', regex=False)
movies_latest['languages']
```

```python
#Simplifying countries of origin
movies_latest['countries_origin'].astype(str)
movies_latest['countries_origin'] = movies_latest['countries_origin'].str.replace('[', '', regex=False).str.replace(']', '', regex=False).str.replace("'", '', regex=False)
movies_latest['countries_origin']
```

```python
#Simplifying writers column
movies_latest['writers'].astype(str)
movies_latest['writers'] = movies_latest['writers'].str.replace('[', '', regex=False).str.replace(']', '', regex=False).str.replace("'", '', regex=False)
movies_latest['writers']
```

```python
#Simplifying stars column
movies_latest['stars'].astype(str)
movies_latest['stars'] = movies_latest['stars'].str.replace('[', '', regex=False).str.replace(']', '', regex=False).str.replace("'", '', regex=False)
movies_latest['stars']
```

```python
#Simplifying directors column
movies_latest['directors'].astype(str)
movies_latest['directors'] = movies_latest['directors'].str.replace('[', '', regex=False).str.replace(']', '', regex=False).str.replace("'", '', regex=False)
movies_latest['directors']
```

```python
movies_latest['MPA'] = movies_latest['MPA'].fillna('Unrated')
movies_latest['MPA']
```

```python
all_genres = movies_latest['genres'].str.split(',').explode()
genre_counts = all_genres.str.strip().value_counts().sort_values(ascending=False).reset_index()

genre_counts = pd.DataFrame(genre_counts)
print(genre_counts['genres'].unique())
```

```python
genre_counts
```

```python
movies_missing = movies_latest[~(movies_latest['description'].isna()) &
                               ~(movies_latest['rating'].isna()) &
                               ~(movies_latest['duration'].isna()) &
                               ~(movies_latest['meta_score'].isna())]

movies_missing = movies_missing.drop(columns=['filming_locations', 'release_date', 'Movie_Link', 'index'])
movies_missing.columns
```

```python
movies_missing
```

```python
movies_missing['words_in_description'] = movies_missing['description'].str.split().str.len() 
movies_missing
```

```python
movies_missing['words_in_description'].describe()
```

```python
movies_missing['tagged_description'] = movies_missing[['id', 'description']].astype(str).agg(' '.join, axis=1)
movies_missing
```

```python
movies_missing.drop(columns=['words_in_description'], axis=1).to_csv('movies_cleaned.csv', index=False)
```
