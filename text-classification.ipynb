```python
subgenre_to_category = {}
for category, subgenres in genre_mapping.items():
    for sub in subgenres:
        subgenre_to_category[sub] = category

def map_to_main_genres(genre_str):
    if pd.isna(genre_str):
        return []
    matched_genres = set()
    for genre in [g.strip() for g in genre_str.split(',')]:
        if genre in subgenre_to_category:
            matched_genres.add(subgenre_to_category[genre])
    return list(matched_genres)

movies['simple_genres'] = movies['genres'].apply(map_to_main_genres)
movies
```

```python
from transformers import pipeline

genres = ['Drama', 'Comedy', 'Thriller', 'Romance', 'Action', 'Documentary', 'Adventure', 'Mystery', 'Horror', 'Fantasy', 'Crime', 'Sci-Fi', 
          'Sport', 'Family', 'Musical', 'Animation']

pipe = pipeline('zero-shot-classification', model='facebook/bart-large-mnli')
```

```python
sequence = movies.loc[0, 'description']
result = pipe(sequence, genres, multi_label=True)
print(result)
```

```python
max_index = np.argmax(result['scores'])
max_label = result['labels'][max_index]
max_label
```

```python
def generate_predictions(sequence, genres):
    predictions = pipe(sequence, genres, multi_label=True)
    max_index = np.argmax(predictions['scores'])
    max_label = predictions['labels'][max_index]
    return max_label
```

```python
from tqdm import tqdm

actual_genres = []
predicted_genres = []

for i in tqdm(range(0, 300)):
    sequence = movies.loc[i, 'description']
    predicted_genres += [generate_predictions(sequence, genres)]

for i in range(0, 300):
    actual = movies.loc[i, 'simple_genres']
    actual_genres.append(actual)
```

```python
predictions_df = pd.DataFrame({'actual_genres': actual_genres, 'predicted_genres': predicted_genres})
predictions_df.head(25)
```
